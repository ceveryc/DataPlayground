# Пример работы (погодная статистика)

Визуализация данных о погоде на метеостанции Краснодара за 10 лет с 2009 по 2018 год включительно.

## 1. Данные для анализа
Открытые данные с сайта rp5.ru (https://rp5.ru/Архив_погоды_в_Краснодаре,_Круглике)

Запросы сохранены локально:
* 34927.01.01.2009.01.01.2014.1.0.0.ru.utf8.00000000.csv - полная статистика за 2009-2013 год (2009 с 25 марта, предыдушие записи не доступны)
* 34927.01.01.2014.01.01.2019.1.0.0.ru.utf8.00000000.csv - полная статистика за 2014-2018 год
* 34927.01.03.2019.31.03.2019.1.0.0.ru.utf8.00000000.xls - полная статистика за март 2019 года

Параметры для анализа:
+ Date - местные дата и время
+ Temp - температура (°C)
+ Pres - давление (мм. ртутного столба)
+ Hum - влажность (%)
+ WinD - направление ветра
+ WinF - средняя сила ветра (м/с)
+ Cloud - общая облачность (%)
+ Prec - количество осадков (мм.)

#### Настройка и подготовка данных:

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
%config InlineBackend.figure_format = 'svg' 
plt.style.use('fivethirtyeight')
plt.rcParams['figure.figsize'] = [10, 5]

columns = [0,1,2,5,6,7,10,23]
names = ['Date','Temp','Pres','Hum','WinD','WinF','Cloud','Prec']

#read first 5 years
first5 = pd.read_csv('data/34927.01.01.2009.01.01.2014.1.0.0.ru.utf8.00000000.csv',
                    sep=';', header=None, skiprows=7, usecols=columns, names=names)
#remove first 8 rows - 01.01.2014 and last 51 - (25-31).03.2009
first5 = first5[8:][:-51]

#read second 5 years
second5 = pd.read_csv('data/34927.01.01.2014.01.01.2019.1.0.0.ru.utf8.00000000.csv',
                    sep=';', header=None, skiprows=7, usecols=columns, names=names)
#remove first 8 rows - 01.01.2019
second5 = second5[8:]

# merge and sort by date
decade = pd.concat([first5, second5])
decade['Date'] = pd.to_datetime(decade.Date, format='%d.%m.%Y %H:%M')
decade = decade.set_index('Date').sort_index()
decade.tail()
```

После подготовки таблица содержит 28656 записей

## 1. Температура по месяцам

```python
y_m_group = decade.Temp.groupby([decade.index.year, decade.index.month])
y_m_min = y_m_group.min()
y_m_max = y_m_group.max()

m_max_max = y_m_max.groupby(level=1).max()
m_max_mean = y_m_max.groupby(level=1).mean()

m_mean = decade.Temp.groupby(decade.index.month).mean()

m_min_mean = y_m_min.groupby(level=1).mean()
m_min_min = y_m_min.groupby(level=1).min()

plt.plot(m_max_max, linestyle='none', marker='+', color='red', label='Абсолютный максимум');
plt.plot(m_max_mean, linestyle='none', marker='o', color='red', label='Средний максимум');
plt.plot(m_mean, linestyle='none', marker='o', color='green', label='Средняя температура');
plt.plot(m_min_mean,linestyle='none', marker='o', color='blue', label='Средний минимум');
plt.plot(m_min_min,linestyle='none', marker='+', color='blue', label='Абсолютный минимум');

month_labels = ['Янв','Фев','Мар','Апр','Май','Июн','Июл','Авг','Сен','Окт','Ноя','Дек']

plt.xticks(range(1,13), labels=month_labels);
plt.yticks(range(-25, 45, 5), labels=[str(i)+'°C' if i%10==0 else '' for i in range(-25, 50, 5)]);

plt.legend(fontsize='small');

plt.title('Температуры Краснодара с 2009 по 2019 год');
```

## 2. Норма осадков и облачность

Количество осадков дано в текстовом формате, чтобы посчитать его сумму примем константы "Осадков нет" за 0, и "Следы осадков" за 0.01.

```python
# mean ammount of precipitation by month
prec = decade.Prec.replace({"Осадков нет":0,"Следы осадков":0.01}).astype(float)
prec_y_m_sum = prec.groupby([decade.index.year, decade.index.month]).sum()
prec_m_mean = prec_y_m_sum.groupby(level=1).mean()

# mean number of days with precipitation by month
prec_y_m_d_sum = prec.groupby([decade.index.year, decade.index.month,  decade.index.day]).sum()
y_m_days_w_prec = (prec_y_m_d_sum > 0).groupby(level=[0,1]).sum()
m_days_w_prec = y_m_days_w_prec.groupby(level=1).mean()

# visualisation
# precipitation mean
fig, axes = plt.subplots(3, 1, figsize=(10,7), sharex=True)
plt.xticks(range(1,13), labels=['Янв','Фев','Мар','Апр','Май','Июн','Июл','Авг','Сен','Окт','Ноя','Дек']);

axes[0].bar(prec_m_mean.index, height=prec_m_mean)
axes[0].set_ylim(-1, prec_m_mean.max())
axes[0].set_title('Среднее количество осадков (мм.)');

for i, v in enumerate(prec_m_mean):
    axes[0].text(i+1, v/2, '{:.1f}'.format(v), color='white', horizontalalignment='center', verticalalignment='center');
    
# days with precipitation 
axes[1].bar(m_days_w_prec.index, height=m_days_w_prec, color='blue')
axes[1].set_ylim(-0.3,m_days_w_prec.max());
axes[1].set_title('Среднее количество дней с осадками');

# clouds
cloud_map = {
    '100%.': 100,
    'Облаков нет.': 0,
    '90  или более, но не 100%': 90,
    '70 – 80%.': 75,
    '20–30%.': 25,
    '60%.': 60,
    '40%.': 40,
    '10%  или менее, но не 0': 10,
    '50%.': 50,
    'Небо не видно из-за тумана и/или других метеорологических явлений.': None
}

cloud = decade.Cloud.replace(cloud_map).dropna()
clouds_mean = cloud.groupby(cloud.index.month).mean()

axes[2].bar(clouds_mean.index, height=clouds_mean);
axes[2].set_title('Средняя облачность (%)');
axes[2].set_ylim(-1,101);

plt.tight_layout()
```

## 3. Роза ветров

Роза ветров будет отображать 3 показателя: 
+ направление (направление от центра спектрограммы)
+ сила ветра (расстрояние от центра спектрограммы)
+ количество измерений каждой пары (цвет)

```python
wind = decade[['WinD', 'WinF']].dropna()
wind = wind.groupby([wind.index.month, 'WinD', 'WinF']).size().reset_index([1,2]).rename(columns={0:'Count'})

direction_map = {
    'Ветер, дующий с запада': 0 * np.pi/8,
    
    'Ветер, дующий с западо-юго-запада': 1 * np.pi/8,
    'Ветер, дующий с юго-запада': 2 * np.pi/8,
    'Ветер, дующий с юго-юго-запада': 3 * np.pi/8,
    
    'Ветер, дующий с юга': 4 * np.pi/8,
    
    'Ветер, дующий с юго-юго-востока': 5 * np.pi/8,
    'Ветер, дующий с юго-востока': 6 * np.pi/8,
    'Ветер, дующий с востоко-юго-востока': 7 * np.pi/8,
    
    'Ветер, дующий с востока': 8 * np.pi/8,
    
    'Ветер, дующий с востоко-северо-востока': 9 * np.pi/8,
    'Ветер, дующий с северо-востока': 10 * np.pi/8,
    'Ветер, дующий с северо-северо-востока': 11 * np.pi/8,
    
    'Ветер, дующий с севера': 12 * np.pi/8,
    
    'Ветер, дующий с северо-северо-запада': 13 * np.pi/8,
    'Ветер, дующий с северо-запада': 14 * np.pi/8,
    'Ветер, дующий с западо-северо-запада': 15 * np.pi/8,
    
    'Штиль, безветрие': -1 
}

titles=['Январь','Февраль','Март','Апрель','Май','Июнь','Июль','Август','Сентябрь','Октябрь','Ноябрь','Декабрь']

fig, axes = plt.subplots(nrows=4, ncols=3, figsize=(10, 14))

def add_compas_mark(ax):
    ax.text(27.0, 26.0, 'N', color='blue', fontsize='x-small');
    ax.text(25.1, 28.0, 'W', color='white', fontsize='x-small');
    ax.text(26.0, 28.8, '✧', color='white', fontsize='xx-large');
    ax.text(28.8, 28.0, 'E', color='white', fontsize='x-small');
    ax.text(27.1, 29.8, 'S', color='red', fontsize='x-small');

for month in range(1,13):

    mid = 15
    Z = np.full((31,31), 1)
    
    df = wind.loc[month]
    
    ax = axes.flat[month-1]
    ax.set_axis_off()
    ax.set_title(titles[month-1], fontsize='medium')
    
    for ind, row in df.iterrows():    
        rad = direction_map[row.WinD]
        x = mid+int(round(row.WinF*np.cos(rad)))
        y = mid-int(round(row.WinF*np.sin(rad)))
        Z[y][x] += row.Count

    im = ax.imshow(Z, cmap='magma', vmax=300, norm=LogNorm(1, 300));
    add_compas_mark(ax)

fig.tight_layout()
```

### 3.1 Роза ветров
Альтернативное отображение с помощью библиотеки seaborn

```python
fig, axes = plt.subplots(nrows=4, ncols=3, figsize=(10, 14))

for month in range(1,13):
    
    df = wind.loc[month]

    x = np.empty(len(df))
    y = np.empty(len(df))
        
    for ind in range(len(df)):
        row = df.iloc[ind]
        rad = direction_map[row.WinD]
        x[ind] = int(round(row.WinF*np.cos(rad)))
        y[ind] = int(round(row.WinF*np.sin(rad)))

    ax = axes.flat[month-1]
    
    sns.kdeplot(x, y, shade=True, ax=ax, cmap='Reds')
    ax.set_aspect('equal', adjustable='box')
    ax.collections[0].set_alpha(0)
    
    ax.set_title(titles[month-1], fontsize='medium')
    ax.set_xlim((-13,13))
    ax.set_ylim((-12,14))
    ax.set_yticks([0])
    ax.set_xticks([0])
    ax.set_yticklabels([])
    ax.set_xticklabels([])
    
    ax.text(0, +12.5, 'N', color='white', horizontalalignment='center', verticalalignment='center', bbox=dict(facecolor='blue'));
    ax.text(0, -10.5, 'S', color='white', horizontalalignment='center', verticalalignment='center', bbox=dict(facecolor='red'));
    ax.text(-11.5, 0, 'W', color='white', horizontalalignment='center', verticalalignment='center', bbox=dict(facecolor='gray'));
    ax.text(+11.5, 0, 'E', color='white', horizontalalignment='center', verticalalignment='center', bbox=dict(facecolor='gray'));
    
fig.tight_layout()
```

Роза ветров составленая с помощью kdeplot реалистичнее отображает средние показатели, но при этом скрывает редкие случаи с большой силой ветра.

## 4.Атмосферное давление

Составим график средних значений давления по месяцам.

```python
y_m_group = decade.Pres.groupby([decade.index.year, decade.index.month])
y_m_min = y_m_group.min()
y_m_max = y_m_group.max()

m_max_mean = y_m_max.groupby(level=1).mean()

m_mean = decade.Pres.groupby(decade.index.month).mean()

m_min_mean = y_m_min.groupby(level=1).mean()

plt.plot(m_max_mean, linestyle='none', marker='o', color='red', label='Средний максимум');
plt.plot(m_mean, linestyle='none', marker='o', color='green', label='Среднее значение');
plt.plot(m_min_mean,linestyle='none', marker='o', color='blue', label='Средний минимум');

plt.xticks(range(1,13), labels=month_labels);
plt.ylabel('мм. ртутного столба')
# plt.yticks(range(-25, 45, 5), labels=[str(i)+'°C' if i%10==0 else '' for i in range(-25, 50, 5)]);

plt.legend(fontsize='small');

plt.title('Атмосферное давление в Краснодаре 2009-2019 год');
```

Из графика видно, что атмосферное давление имеет сезонные колебания, обратные колебаниям температуры.

Изучим зависимость между температурой и давлением. Чтобы исключить влияние времени года, сделаем случайную выборку пар этих параметров.

```python
temp_pres = decade[['Temp', 'Pres']].dropna()

ind = np.random.choice(np.arange(len(temp_pres)), size=2000)

sample = temp_pres.iloc[ind]

cor = np.corrcoef(np.transpose(sample.values))
print('Коэффициент корреляции: {:.3f}'.format(cor[0,1]))

sample.plot(kind='scatter', x='Temp', y='Pres', alpha=0.5);
# plt.title('Зависимость между температурой и давлением')
plt.xlabel('Температура (°C)');
plt.ylabel('Атмосферное давление (мм)');
```

Видно что верхняя граница расположена вдоль прямой линии. Проверим как изменяется граница 95%-интервала давления в зависимости от температуры, для этого сгруппируем измерения с шагом в 2 градуса.

```python
step = 2

trange = np.arange(-18, 38, step)
ar = np.empty((len(trange), 5))

for index, t in np.ndenumerate(trange):
    pres = temp_pres.Pres[(temp_pres.Temp >= (t-step/2)) & (temp_pres.Temp < (t+step/2))]
    ar[index] = np.percentile(pres, [2.5, 25, 50, 75, 97.5])

plt.fill_between(trange, ar[:,0], ar[:,4], color='darkgreen', alpha=0.5, label='95%-интервал');
plt.fill_between(trange, ar[:,1], ar[:,3], color='darkgreen', label='50%-интервал');

a, b = np.polyfit(trange, ar[:,4], 1)
x = np.array([trange.min(), trange.max()])
y = x * a + b
plt.plot(x, y, color='darkgreen', linewidth=1, linestyle='--')

plt.xlim(x);

plt.legend()

plt.xlabel('Температура (°C)');
plt.ylabel('Атмосферное давление (мм)');
```

```python
cor = np.corrcoef(trange, ar[:,4])
print('Коэффициент корреляции верхней границы равен {:.3f}.'.format(cor[0,1]))

index_of_0 = np.where(trange == 0)[0][0]

r_min = 1
r_min_temp = 0
r_min_index = 0
for i in range(10):
    index = index_of_0 + i
    cor = np.corrcoef(trange[:index], ar[:index,0])
#     print('t < {:.1f} : r = {:.3f}'.format(, cor[0,1]))
    if (cor[0,1] < r_min):
        r_min = cor[0,1]
        r_min_temp = trange[index]
        r_min_index = index

print('\nКоэффициент корреляции нижней границы ближе всего к -1 на участке до {:.0f}°C равен {:.3f}.'.format(r_min_temp, r_min))

cor = np.corrcoef(trange[r_min_index:], ar[r_min_index:,0])
print('- на участке от {:.0f}°C равен {:.3f}.'.format(r_min_temp, cor[0,1]))
```
